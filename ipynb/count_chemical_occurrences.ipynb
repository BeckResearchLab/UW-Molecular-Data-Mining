```python
## Download from web of science
## Using chemextractor to extract chemical names from the paper and count the occurrences
```

```python
from chemdataextractor import Document
import pandas as pd
import os
from bs4 import BeautifulSoup
from chemdataextractor.doc import Document, Heading, Paragraph
from collections import Counter
```

```python
chemical_list_fr = []
os.chdir('..')
for filename in os.listdir('./data/flame_retardants/'):
    if filename.endswith(".html"): 
        print(filename)
        f = open('./data/flame_retardants/'+filename,'rb')
        contents = f.read()
        soup = BeautifulSoup(contents)
        text = soup.get_text()
        doc = Document(
            Heading(text)
            )
        for i in range(len(doc.cems)):
            items = []
            items.append(str(doc.cems[i]))
            unique_items= list( dict.fromkeys(items) )
            chemical_list_fr.extend(unique_items)
```

```python
chemical_list_fr
```

```python
counts = Counter(chemical_list_fr)
type(counts)
flame_retardant_list = pd.DataFrame.from_dict(counts, orient='index').sort_values(by = [0], ascending=False)
flame_retardant_list
```

```python
flame_retardant_list.to_csv('./flame_retardant_count.csv')
```
